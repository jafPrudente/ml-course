## Estimación del valor de $\pi$ con el método de Montecarlo

Consideremos un cuadrado de base 1 y altura 1, también un cuarto de circunferencia inscrito en este cuadrado y con centro en el vértice de abajo a la izquierda del cuadrado.

Así mismo denotemos el área del cuarto de circunferencia por $ A_{\circ} $ y el área del cuadrado por $ A_{\square} $.

Dada una pareja $ (x, y) $ diremos que pertenece al área $ A_{\circ} $ si se cumple que $ x^2 + y^2 \leq 1 $; mientras que si $ x^2 + y^2 > 1 $ entonces no pertenece a esta área. 

La probabilidad de que esta pareja $ (x,y) $ pertenezca a $ A_{\circ} $ la podemos escribir como:

$$
    P(\vec{x} \in A_{\circ}) = \dfrac{A_{\circ}}{A_{\square}} = \dfrac{\pi}{4}
$$

Porque $ A_{\square} = 1 $ y $ A_{\circ} = \pi/4 $ (al ser un cuarto de circunferencia). Si generamos $ N $ parejas también podemos aproximar esta probabilidad por:

$$
    P(\vec{x} \in A_{\circ}) \approx \dfrac{n}{N}
$$

donde $ n $ denota a los casos favorables (las parejas dentro de $ A_{\circ} $). Entonces podemos escribir a $ \pi $ de forma aproximada como:

$$
    \pi \approx \dfrac{4 n}{N}
$$

Entonces generemos $ N $ parejas en $ [0,1] \times [0,1] $ y denotemos por $ n $ a todas las que esten dentro del cuarto de circunferencia, y así al calcular el cociente de arriba podremos dar una estimación de $ \pi $

```python
# Importamos bibliotecas.
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
```

```python
def main(N):

    # Generamos la pareja aleatoría con distribución normal.
    x = rd.uniform(0, 1, N)
    y = rd.uniform(0, 1, N)

    # Definimos un contador de casos favorables y contamos estos casos.
    n = 0
    for i in range(N):
        if x[i]**2 + y[i]**2 <= 1:
            n += 1

    # Calculamos la aproximación de pi.
    pi = (4*n)/N
    
    # Graficamos.
    plt.scatter(x, y)
    t = np.linspace(0, 1, 200)
    plt.plot(t, np.sqrt(1 - t**2), color='red')
    plt.title(rf'$\pi \approx {pi}$')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$y$')
    
    return
```

```python
main(1000)
```
