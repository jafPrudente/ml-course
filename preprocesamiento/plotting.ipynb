```python
# Importamos bibliotecas.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos rutas de archivos.
datasets = '../datasets/'
name = 'customer-churn-model/Customer Churn Model.txt'
path = datasets + name
```

```python
# Definimos el dataframe.
df = pd.read_csv(path)
df.head()
```

```python
# Graficamos.
# Podemos usar matplotlib.
plt.scatter(df['Night Mins'], df['Night Charge'])

# O También podemos usar el método plot.
#df.plot(kind='scatter', x='Day Mins', y='Day Charge')
```

```python
# Hagamos otra gráfica con matplotlib.
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)

axs[0, 0].scatter(df['Day Mins'], df['Day Charge'])
axs[0, 0].set_xlabel(r'Day Mins')
axs[0, 0].set_ylabel(r'Day Charge')

axs[0, 1].scatter(df['Night Mins'], df['Night Charge'])
axs[0, 1].set_xlabel(r'Night Mins')
axs[0, 1].set_ylabel(r'Night Charge')

axs[1, 0].scatter(df['Day Calls'], df['Day Charge'])
axs[1, 0].set_xlabel(r'Day Calls')
axs[1, 0].set_ylabel(r'Day Charge')

axs[1, 1].scatter(df['Night Calls'], df['Night Charge'])
axs[1, 1].set_xlabel(r'Night Calls')
axs[1, 1].set_ylabel(r'Night Charge')
```

```python
# Ahora hagamos un histograma.

# Para definir el número de bins se usa la regla de Sturges.
sturges = 1 + np.log2(len(df['Day Calls'].tolist()))
plt.hist(df['Day Calls'], bins = round(sturges))
plt.xlabel(r'Llamadas al día')
```
