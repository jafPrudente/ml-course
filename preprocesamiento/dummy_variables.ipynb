```python
# Importamos bibliotecas.
import pandas as pd
```

```python
# Definimos rutas de archivos.
datasets = '../datasets/'
name = 'titanic/titanic3.csv'
path = datasets + name
```

```python
# Definimos el dataframe.
df = pd.read_csv(path)
```

```python
# Para definir dummy variables se hace lo siguiente.
dummy_sex = pd.get_dummies(df['sex'], prefix='sex')
dummy_sex.head()
```

```python
# En dado caso de que queramos un valor numérico podemos hacer lo siguiente.
dummy_sex = pd.get_dummies(df['sex'], prefix='sex', dtype=int)
dummy_sex.head()
```

```python
# Para cambiar la columna 'sex' por estas dos columnas dummies hacemos lo siguiente.

# Eliminamos la columna 'sex'.
df = df.drop(['sex'], axis=1)

# Concatenamos el dataframe con las variables dummies que definimos anteriormente.
df = pd.concat([df, dummy_sex], axis=1)
df.head()
```

```python
# Todo este proceso lo podemos hacer con una función.
def dummy_vars(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name, dtype=int)
    df = df.drop([var_name], axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```

```python
# Probamos la función
df = pd.read_csv(path)
df = dummy_vars(df, 'sex')
df.head()
```
