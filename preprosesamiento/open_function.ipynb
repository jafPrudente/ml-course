```python
# Importamos bibliotecas.
import numpy as np
import pandas as pd
```

```python
# Abrimos el dataframe.
datasets_path = '../datasets/customer-churn-model/'
file_name = 'Customer Churn Model.txt'
file_path = datasets_path + file_name

# Abrimos en modo de lectura, 'r'.
file = open(file_path, 'r')
```

```python
# Con 'readline' leemos la primer línea, con 'strip' eliminamos espacios en blanco, y con 'split' dividimos por comas.
cols = file.readline().strip().split(',')
print(cols)
```

```python
# Podemos crear un diccionario 'dd' (data dictionary) que sea equivalente a un dataframe de pandas con lo siguiente.
dd = {}
for col in cols:
    dd[col] = []

# Definimos un contador de filas y uno de columnas.
num_lines = 0
num_cols = len(cols)

# Lo rellenamos barriendo el archivo.
for line in file:
    values = line.strip().split(',')
    for i in range(num_cols):
        dd[cols[i]].append(values[i])
    num_lines += 1

print(f'Tenemos {num_cols} columnas, y {num_lines} filas')

# Cerramos el archivo que abrimos.
file.close()
```

```python
# Con pandas podemos crear un dataframe con este diccionario.
df = pd.DataFrame(dd)
df.head()
```
